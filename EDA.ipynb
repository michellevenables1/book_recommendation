```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from collections import Counter
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS 
```

```python
data = pd.read_csv('df3.csv', index_col = 'Unnamed: 0')
```

```python
data.shape
```

```python
# get count of different beers in each style
book_counts = {}
for x in list(data['title']):
    if not x in book_counts:
        book_counts[x] = 1
    else:
        book_counts[x] += 1

author_counts = {}
for x in list(data['author']):
    if not x in author_counts:
        author_counts[x] = 1
    else:
        author_counts[x] += 1

publisher_counts = {}
for x in list(data['publisher']):
    if not x in publisher_counts:
        publisher_counts[x] = 1
    else:
        publisher_counts[x] += 1

state_counts = {}
for x in list(data['state']):
    if not x in state_counts:
        state_counts[x] = 1
    else:
        state_counts[x] += 1
```

```python
# top ten books and counts
top10_books = dict(Counter(book_counts).most_common(10))
top10_authors = dict(Counter(author_counts).most_common(10))
top10_publishers = dict(Counter(publisher_counts).most_common(10))
top10_states = dict(Counter(state_counts).most_common(10))
```

```python
#flipping the count on a horizontal bar chart so the top rated are on the top
top10_books = dict(sorted(Counter(book_counts).most_common(10), key=lambda x:x[1]))
top10_authors = dict(sorted(Counter(author_counts).most_common(10), key=lambda x:x[1]))
top10_publishers = dict(sorted(Counter(publisher_counts).most_common(10), key=lambda x:x[1]))
top10_states = dict(sorted(Counter(state_counts).most_common(10), key=lambda x:x[1]))
```

```python
# Plots most popular book ratings 
fig = plt.figure(figsize = (8,4))
plt.barh(range(len(top10_books)), list(top10_books.values()), align='center')
plt.yticks(range(len(top10_books)), list(top10_books.keys()), fontsize=12)
plt.title("Most Popular Book Ratings", fontsize=16, fontweight= 22)
plt.grid(b=None)
mpl.rcParams.update({'text.color' : "darkblue",
                     'axes.labelcolor' : "black",
                     'xtick.color' : "black",
                     'ytick.color' : "black"})
plt.tight_layout()
plt.show()
fig.savefig('mostpopbook.png', transparent=True, edgecolor='none')
```
